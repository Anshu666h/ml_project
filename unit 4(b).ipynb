```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sklearn.datasets

data = pd.read_csv('/content/bank.csv')


print(data.columns)


target_column = 'age'


correlation = data['age'].corr(data[target_column])
print(f'Correlation between age and {target_column}: {correlation}')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y=target_column, data=data)
plt.title(f'Scatter Plot of age vs {target_column}')
plt.xlabel('age')  # Replace 'charges' here as well
plt.ylabel(target_column)
plt.grid(True)
plt.show()
```
